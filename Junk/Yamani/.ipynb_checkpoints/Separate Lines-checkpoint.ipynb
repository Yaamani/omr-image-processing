```python
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
from skimage import io
from skimage.color import *
import cv2
from my_commonfunctions import *

#matplotlib.rcParams['figure.dpi'] = 200
```

```python
def close(src, kernel):
    dilated = cv2.dilate(src, kernel)
    return cv2.erode(dilated, kernel)
```

```python
test3_gray = (rgb2gray(rgba2rgb(io.imread('test3.png'))) * 255).astype(np.uint8)
test3_binary = test3_gray < 200
test3_thresholded_gray = gray2rgb(test3_binary * 255).astype(np.uint8)
#print(test3_thresholded_gray.dtype)
```

```python
edges = cv2.Canny(test3_gray, 100, 200)
kc1 = np.ones((7, 7))
closed1 = close(edges, kc1)
my_show_images([test3_gray, edges, closed1], row_max=2)
```

```python
closed1_bin = closed1 > 200
is_line = (np.sum(closed1_bin, axis=1) / test3_gray.shape[1]) > 0.6
# is_line
image, contours, hierarchy = cv2.findContours((is_line*255).astype(np.uint8), 
                                              cv2.RETR_TREE, 
                                              cv2.CHAIN_APPROX_SIMPLE)
```

```python
contours = np.array(contours).squeeze()
contours_y = contours.T[1].T
contour_mid_point = np.sort(np.average(contours_y, axis=1))
contour_mid_point
```
